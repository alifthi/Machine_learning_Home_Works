```python
data = data.to_numpy()
np.random.shuffle(data)
test_size = int(data.shape[0]*test_size)
x_test = data[:test_size,2:-1]
y_test=data[:test_size,1]
x_train = data[test_size:,2:-1]
y_train=data[test_size:,1]
x_train.shape
```

```python
means_M = np.mean(x_train[y_train=='M'],axis=0)
std_M = np.std(x_train[y_train=='M'].astype(np.float64),axis=0)
means_B = np.mean(x_train[y_train=='B'],axis=0)
std_B = np.std(x_train[y_train=='B'].astype(np.float64),axis=0)
num_of_data=x_train.shape[0]
# num_of_m = y_train[targets=='M'].shape[0]
num_of_m=np.count_nonzero(y_train=='M')
num_of_b = num_of_data-num_of_m
p_m = num_of_m/num_of_data
p_b = num_of_b/num_of_data
```

```python
print(f'number of class M: {num_of_m}, P(M)={p_m}')
print(f'number of class B: {num_of_b}, P(B)={p_b}')
```

```python
f_g = lambda x,s,m:np.exp(-0.5*(((x-m)/s)**2))/(s*np.sqrt(2*np.pi))
```

```python
f_g(1,2,3)
```

```python
def predict(data):
    classes={0:'class_M',1:'class_B'}
    preds_if_M=1
    preds_if_B=1
    for i in range(len(means_M)):
        preds_if_M = preds_if_M*f_g(x=data[i],s=std_M[i],m=means_M[i])
        preds_if_B = preds_if_B*f_g(x=data[i],s=std_B[i],m=means_B[i])
    confid = preds_if_M*p_m - preds_if_B*p_b
    return classes[int(confid<0)],confid
```

```python
idx = 0
print(f'predicted class: {predict(x_test[idx])[0]}, label: {y_test[idx]}')
```

```python
pred=[]
confidence=[]
for sample in x_test:
    predinction,confid=predict(sample)
    pred.append(predinction[-1])
    confidence.append(confid)
```

```python
prediction_dict={'MM':0,'MB':1,'BM':2,'BB':3}
confusion_matrix=[0,0,0,0]
confidence_matrix=[0,0,0,0]
for i,p in enumerate(pred):
    section = prediction_dict[p+y_test[i]]
    confusion_matrix[section]+=1
    confidence_matrix[section]+=confidence[i]
for i in range(4):
    confidence_matrix[i]/=confusion_matrix[i]
```

```python
print(f'confusion matrix: {np.reshape(confusion_matrix,[2,2])}')
print(f'confidence matrix: {np.reshape(confidence_matrix,[2,2])}')
```

```python
print(f'precision: {confusion_matrix[0]/(confusion_matrix[0]+confusion_matrix[1])}')
print(f'recall: {confusion_matrix[0]/(confusion_matrix[0]+confusion_matrix[2])}')
print(f'Accuracy: {(confusion_matrix[0]+confusion_matrix[3])/sum(confusion_matrix)}')
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics 
from sklearn.metrics import confusion_matrix
gnb = GaussianNB()
y_train[y_train=='M'] = 0
y_train[y_train=='B'] = 1
y_train = list(y_train)
pred = gnb.fit(x_train, y_train).predict(x_test)
y_test[y_test=='M'] = 0
y_test[y_test=='B'] = 1
mat=confusion_matrix(y_true=list(y_test),y_pred=pred)
precision=metrics.precision_score(pred,list(y_test))
recall=metrics.recall_score(pred,list(y_test))
print(f'confusion matrix: {mat}')
print(f'precision: {precision}')
print(f'recall: {recall}')
```
