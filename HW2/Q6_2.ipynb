```python
import numpy as np
from sklearn.datasets import load_iris
from Q6_1_KNN import KNN
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
test_size=0.2
```

```python
iris=load_iris()
```

```python
data=iris['data']
targets=iris['target']
```

```python
samples=pd.DataFrame(data,columns=iris['feature_names'])
```

Q6_4

```python
train_data,test_data,train_targets,test_targets=train_test_split(data,targets,test_size=test_size)
```

Q6_5

```python
model=KNN(K=5)
model.fit(train_data,train_targets)
prediction=model.transform(test_data)
score=accuracy_score(test_targets,prediction)
print(f'Accuracy for test data: {score}')
```

```python
prediction=model.transform(train_data)
score=accuracy_score(train_targets,prediction)
print(f'Accuracy for train data: {score}')
```

```python
train_scores=[]
test_scores=[]
for k in range(10):
    model=KNN(K=k)
    model.fit(train_data,train_targets)
    prediction=model.transform(test_data)
    test_scores.append(accuracy_score(test_targets,prediction))
    
    prediction=model.transform(train_data)
    print(prediction)
    train_scores.append(accuracy_score(train_targets,prediction))
plt.plot(test_scores)
plt.plot(train_scores)
plt.legend(['test score','train score'])
```
