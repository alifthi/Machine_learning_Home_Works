```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N=1000
```

```python
np.random.seed(1)
X = np.concatenate((np.random.normal(0,1,int(0.3*N)),np.random.normal(5,1,int(0.7*N))))[:,np.newaxis]
```

```python
gussian_func=lambda x: np.exp(-x**2)/np.sqrt(2*np.pi)
```

```python
f_x0=lambda x0,h: sum([gussian_func((x0-x)/h) for x in X])/(N*h)
```

```python
for h in [0.1,1,10]:
    dist=[]
    for x in np.arange(-3,8,0.1):
        dist.append(f_x0(x,h))
    fig = plt.figure(num=None, figsize=(12,8), dpi=80, facecolor='w', edgecolor='k')
    plt.subplot(1, 2, 1)
    plt.plot(np.arange(-3,8,0.1),dist)
    plt.title(f'h= {h}')
    plt.subplot(1, 2, 2)
    plt.hist(X,50)
    plt.show()
    
```
